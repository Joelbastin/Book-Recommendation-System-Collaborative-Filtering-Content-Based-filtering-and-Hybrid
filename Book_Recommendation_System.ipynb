Before moving into processing the text data some basic text preprocessing work has been done

```python
import re
import string
from tqdm.auto import tqdm
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')

from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk import word_tokenize
```

```python
# some text cleaning functions
def convert_to_lower(text):
    return text.lower()

def remove_numbers(text):
    number_pattern = r'\d+'
    without_number = re.sub(pattern=number_pattern, repl=" ", string=text)
    return without_number

def lemmatizing(text):
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    for i in range(len(tokens)):
        lemma_word = lemmatizer.lemmatize(tokens[i])
        tokens[i] = lemma_word
    return " ".join(tokens)

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_stopwords(text):
    removed = []
    stop_words = list(stopwords.words("english"))
    tokens = word_tokenize(text)
    for i in range(len(tokens)):
        if tokens[i] not in stop_words:
            removed.append(tokens[i])
    return " ".join(removed)

def remove_extra_white_spaces(text):
    single_char_pattern = r'\s+[a-zA-Z]\s+'
    without_sc = re.sub(pattern=single_char_pattern, repl=" ", string=text)
    return without_sc
```

```python
df5['Des'] = df5['Description'].apply(lambda x: convert_to_lower(x))
df5['Des'] = df5['Description'].apply(lambda x: remove_numbers(x))
df5['Des'] = df5['Description'].apply(lambda x: remove_punctuation(x))
df5['Des'] = df5['Description'].apply(lambda x: remove_stopwords(x))
df5['Des'] = df5['Description'].apply(lambda x: remove_extra_white_spaces(x))
df5['Des'] = df5['Description'].apply(lambda x: lemmatizing(x))
```

### Hybrid Recommendation

For Hybrid, we have to use the data which is common to both model. So inner merge is performed. And have to train everytihng again to avoid facing "out of bound error" because shape of the latest dataframe is changed.

```python
# Collaborative Filtering
def COLAB(Book_title,mod = mod):
  csr1 = csr_matrix(pivot1.values)
  mod1 = NearestNeighbors(metric = "cosine")
  mod1.fit(csr1)
  ind = pivot1.index.get_loc(Book_title)
  distance, indices = mod1.kneighbors(pivot1.iloc[ind,:].values.reshape(1,-1), n_neighbors= 100 )

  colab = pd.DataFrame()

  for i in range(1,50):
      b = pivot1.index[indices[0, i]]
      colab = pd.concat([colab,pd.DataFrame({'Title': [b]})], ignore_index=True)
  return colab


# Content based Filtering
def CONTENT(title, cs = cs):
    vector1 = cv.fit_transform(df6["Des"]).toarray()
    cs1 = cosine_similarity(vector1)


    indices = pd.Series(df6.index, index = df6["original_title"])
    content = pd.DataFrame()

    indx = indices[title]

    score = list(enumerate(cs1[indx]))

    scores = sorted(score, key = lambda x: x[1], reverse = True)

    scores = scores[1:50]

    movie_indices = [i[0] for i in scores]
    d = df6[["original_title"]].iloc[movie_indices]
    d = pd.DataFrame(d["original_title"].unique(), columns = ["Title"])

    return d

def Hybrid_Recommendation(book_title):
  col = COLAB(book_title)
  con = CONTENT(book_title)
  rec = pd.merge(col, con, how = "inner", on = "Title")
  return rec.iloc[:10,:]
```
